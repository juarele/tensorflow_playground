## Create the data

```python
import numpy as np

num_puntos = 1000
conjunto_puntos = []
for i in xrange(num_puntos):
    # create random variable, x1, with zero-mean and 0.55 std
    x1 = np.random.normal(0.0,0.55)
    # create dependent variable y = 0.1*x+0.3+noise
    y1 = x1*0.1+0.3+np.random.normal(0.0,0.03)
    conjunto_puntos.append([x1,y1])

x_data = [v[0] for v in conjunto_puntos]
y_data = [v[1] for v in conjunto_puntos]
```

## Plot the data

```python
%matplotlib inline

import matplotlib.pyplot as plt

plt.plot(x_data,y_data,'ro',label = 'Original data')
plt.legend()
plt.show()
```

## Define regresison vars with TensorFlow

```python
import tensorflow as tf
```

```python
W = tf.Variable(tf.random_uniform([1],-1,1))
```

```python
b = tf.Variable(tf.zeros([1]))
```

```python
y = W*x_data+b
```

```python
loss = tf.reduce_mean(tf.square(y-y_data))
```

```python
optimizer = tf.train.GradientDescentOptimizer(0.6)
train = optimizer.minimize(loss)
```

## Execution

```python
session = tf.Session()
session.run(tf.initialize_all_variables())
```

```python
for step in xrange(16):
    session.run(train)
    print step,session.run(W),session.run(b)
    plt.figure(figsize=(8, 4), dpi=80, facecolor='w', edgecolor='k')
    plt.plot(x_data,y_data,'ro',label = 'Original data')
    plt.plot(x_data,session.run(W)*x_data+session.run(b),'-b',label = 'Linear regression')
    plt.legend(loc=0)
    plt.show()
```

```python
plt.figure(figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.plot(x_data,y_data,'ro',label = 'Original data')
plt.plot(x_data,session.run(W)*x_data+session.run(b),'-b',label = 'Linear regression')
plt.legend(loc=0)
plt.show()
```
